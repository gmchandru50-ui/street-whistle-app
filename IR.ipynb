```python
# Porter Stemmer Implementation
# Morphological analyzer that reduces words to their grammatical roots

import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download tokenizer dataset (only first time)
nltk.download('punkt')
nltk.download('punkt_tab')

# Create Porter Stemmer object
ps = PorterStemmer()

# Take input from user
sentence = input("Enter a sentence: ")

# Tokenize sentence into words
words = word_tokenize(sentence)

print("\nOriginal Words -> Stemmed Words\n")

# Apply stemming
for word in words:
    stem = ps.stem(word)
    print(word, " -> ", stem)
```
